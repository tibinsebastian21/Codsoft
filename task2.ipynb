```python
import pandas as pd

data = pd.read_csv('ind_movies.csv', encoding='latin-1')
data.head(10)
```

```python
data.isnull().sum()
```

```python
data['Votes'] = data['Votes'].str.replace(',', '').str.replace('M', '000000').str.replace('.','').str.replace('$','').astype(float)
data['Votes'] = pd.to_numeric(data['Votes'], errors='coerce')
data['Votes'].fillna(data['Votes'].median(), inplace=True)
data['Rating'].fillna(data['Rating'].median(), inplace=True)
data.dropna(subset=['Name', 'Year', 'Duration', 'Genre', 'Director', 'Actor 1', 'Actor 2', 'Actor 3'], inplace=True)
data['Year'] = data['Year'].str.extract('(\d{4})').astype(int)
data['Duration'] = data['Duration'].str.replace(' min', '').astype(int)
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder

categorical_columns = ['Genre', 'Director', 'Actor 1', 'Actor 2', 'Actor 3']
encoded_data = pd.get_dummies(data, columns=categorical_columns)
```

```python
from sklearn.model_selection import train_test_split

x = encoded_data.drop('Rating', axis=1)
y = encoded_data['Rating']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```
